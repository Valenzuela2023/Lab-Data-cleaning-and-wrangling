```python
# Primero quitamos los nan
df_ej_1=df.copy()
df_ej_1=df_ej_1[df_ej_1.notna().all(axis=1)]
df_ej_1
```

```python
# Para quitar los outliers:
# Calcular los percentiles para definir el rango
percentil_25 = df_ej_1['INCOME'].quantile(0.25)
percentil_75 = df_ej_1['INCOME'].quantile(0.75)

# Calcular el rango intercuartÃ­lico (IQR)
IQR = percentil_75 - percentil_25

# Definir los lÃ­mites inferior y superior 
limite_inferior = percentil_25 - 0.5 * IQR
limite_superior = percentil_75 + 0.5 * IQR

# Eliminar los valores atÃ­picos
df_sin_outliers = df_ej_1[(df_ej_1['INCOME'] >= limite_inferior) & (df_ej_1['INCOME'] <= limite_superior)]

print(df_sin_outliers)
```

```python
# Si metemos lo anterior en una funcion:
def remove_outliers_iqr(df_ej_1, INCOME):
    Q1 = df_ej_1[INCOME].quantile(0.25)
    Q3 = df_ej_1[INCOME].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 0.5 * IQR
    upper_bound = Q3 + 0.5 * IQR
    df_ej_1 = df_ej_1[(df_ej_1[INCOME] >= lower_bound) & (df_ej_1[INCOME] <= upper_bound)]
    return df_ej_1
```

```python
df_sin_outliers=remove_outliers_iqr(df_ej_1, "INCOME")
print(df_sin_outliers)
```
